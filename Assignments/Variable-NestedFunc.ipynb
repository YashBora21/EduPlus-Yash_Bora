```python
# 4 Create two functions where both use a global variable to calculate and store cumulative total sales
total=0
def first_sales(*items):
    global total
    
    total+=sum(items)
    print("\nthe total sales for first_sales is :",total)

def second_sales(*items):
    global total
    total+=sum(items)
    print("\nthe total sales for second_sales is :",total)
first_sales(10,203,103)
second_sales(30,20,90)

```

```python
# 6 Demonstrate the use of nonlocal keyword by modifying an enclosing variable inside an inner function
def outer():
    x="Enclosing var"
    def inner():
        nonlocal x
        x="inner var"
        
    print("\n outer function before inner ;",x)
    inner()
    print("\n outer function after inner :",x)

outer()
```

```python
# 13 Demonstrate how enclosing variable stores the running total for a counter function using nonlocal

def outer():
    z=4
    print("\nbefore counter the z is :",z)

    def counter(num):
        nonlocal z
        for i in range(1,num):
            z+=1
        print("\nAfter counter the z is :",z)

    counter(10)
outer()


```

```python
# 24 Demonstrate how a nonlocal variable can be used to maintain the count of how many times a nested function has been called
def outer():
    cnt=0
    def inner():
        nonlocal cnt
        cnt+=1
    inner()
    inner()
    print("\nthe function call this much time: ",cnt)
outer()
```

```python
# 26 Create a nested function example where inner function modifies enclosing variable to accumulate running total of expenses
def expenses():
    totalexpenses=0
    def calaulate(*amount):
        nonlocal totalexpenses
        totalexpenses=sum(amount)
        return totalexpenses
    return calaulate
exp=expenses()
print("the total expenses are :",exp(1000,2030,4000))


```

```python
# 29 Write a function with both *args and **kwargs to simulate order details containing items and their prices

def order(*args,**kwargs):
    print("\n-------order summary-------\n")
    for i in args:
        print(f"\n{i}")
    print("\nPrices:\n")
    total=0
    for key,value in kwargs.items():
        print(f"\n{key}={value}")
        total+=value
    print("\ntotal amount is:",total)


order("laptop","phone","mouse",laptop=50000,phone=10000,mouse=1000)
```
