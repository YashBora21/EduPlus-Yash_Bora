```python
# 12. Create a function that accepts text input and returns a list of nouns and verbs separately using POS tagging

def por():
    noun=[]
    verbs=[]
    text=input("Enter the text: ")
    words=word_tokenize(text)
    protag=nltk.pos_tag(words)
    for word,tag in protag:
        if tag.startswith('NN'):
            noun.append(word)
        elif tag.startswith('VB'):
            verbs.append(word)

    return [noun,verbs]

print(por())
    
```

```python
# 13. Count the number of adjectives and adverbs present in a given text.
def por():
    adjectives=0
    adverbs=0
    text=input("Enter the text: ")
    words=word_tokenize(text)
    protag=nltk.pos_tag(words)
    for word,tag in protag:
        if tag.startswith('NN'):
            adjectives+=1
        elif tag.startswith('RB'):
            adverbs+=1

    return [adjectives,adverbs]

ans=por()

print(f"the count of the adjective is :{ans[0]} the count of the adverbs is :{ans[1]}")
```

```python
# 14. Tokenize a paragraph and find the average length of words using NLTK functions.
def avg_lenof_words():
    text=input("enter the text ")
    words=word_tokenize(text)
    count=0
    count_list=[]
    for i in words:
        count_list.append(len(i))
        count+=1
    avg=sum(count_list)/count
    return avg
print("the average len of word in txt is :",avg_lenof_words())
```

```python
# 21. Load the NLTK movie_reviews dataset and count the number of positive and negative reviews.
import nltk
from nltk.corpus import movie_reviews

# Download dataset (run once)
nltk.download('movie_reviews')

# Count positive and negative reviews
pos_count = len(movie_reviews.fileids('pos'))
neg_count = len(movie_reviews.fileids('neg'))

print("Positive reviews:", pos_count)
print("Negative reviews:", neg_count)
```

```python
# 22. Extract the top 2000 frequent words from the movie_reviews corpus and print them.


all_words=nltk.FreqDist(w.lower() for w in movie_reviews.words()) 
word_freq=list(all_words)[:2000]
word_freq
```

```python
# 23. Create a function that checks whether specific words exist in a review text and returns True/False.
def exist(txt):
    if txt in all_words:
        return True
    else:
        return False
    
print('is really exist in movies reviwes :',exist('really'))
print('is eally exist in movies reviwes :',exist('eally'))
```

```python
# 24. Prepare feature sets using the movie_reviews dataset and split them into training and testing data
documnets=[(list(movie_reviews.words(fileid)),category)
           for category in movie_reviews.categories()
           for fileid in movie_reviews.fileids(category)
           ]
all_words=nltk.FreqDist(w.lower() for w in movie_reviews.words()) 
word_freq=list(all_words)[:2000]
def find_features(documnets):
    words=set(documnets)
    features={}
    for i in word_freq:
        features[i]=(i in words)

    return features

feature_set=[(find_features(rev),category) for (rev,category)in documnets]
train_set,test_set=feature_set[:1600],feature_set[1600:]
print('traning_set:',len(train_set))
print('testing_set:',len(test_set))
```

```python
# 25. Train a Naive Bayes classifier using the movie_reviews dataset and print the accuracy.
classifier=nltk.NaiveBayesClassifier.train(train_set)
accuracy=nltk.classify.accuracy(classifier,test_set)*100
print("the classifier accuracy is: ",accuracy)
```

```python
# 26. Display the top 10 most informative features of the Naive Bayes classifier.
print('top 10 informative info is :',classifier.most_informative_features(10))
```

```python
# 27. Write a function that takes a custom sentence and predicts if it is positive or negative using the trained model

def pos_or_neg(snt):
    tokens = word_tokenize(snt.lower())
    features = find_features(tokens)
    return classifier.classify(features)

print(pos_or_neg('This movie was amazing and fantastic'))
print(pos_or_neg("The film was boring and terrible"))
```

```python
# 28. Save the trained Naive Bayes classifier using pickle and load it again to make predictions.
import pickle

# Save trained classifier
with open("sentiment_model.pkl", "wb") as f:
    pickle.dump(classifier, f)
# Load classifier
with open("sentiment_model.pkl", "rb") as f:
    loaded_classifier = pickle.load(f)

# Test prediction
print(loaded_classifier.classify(find_features(["good", "movie"])))

```

```python
#29. Text Preprocessing: Tokenization, Stopword Removal, Lemmatization
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re


def preprocess_text(text):
    # Lowercase
    text = text.lower()
    
    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    
    # Tokenization
    tokens = word_tokenize(text)
    
    # Stopword removal
    stop_words = set(stopwords.words('english'))
    tokens = [w for w in tokens if w not in stop_words]
    
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(w) for w in tokens]
    
    return tokens

# Example
print(preprocess_text("The movies were better than expected!"))
```
