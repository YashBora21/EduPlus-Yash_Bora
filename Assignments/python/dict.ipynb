```python
## 1. Create a dictionary with 5 students and their marks.
dict_stu={"ajay":89,"jay":78,"vijay":56,"sanjay":45,"yash":82}
print(dict_stu)
```

```python
## 2. Print the value of a specific student using their name as key.
print("the marks of yash is :",dict_stu["yash"])
```

```python
# 3. Add a new student with marks to the dictionary.
dict_stu["ram"]=90
print(dict_stu)
```

```python
# 4. Update the marks of an existing student.
dict_stu["yash"]=90

print(dict_stu)
```

```python
# 5. Delete a student from the dictionary.
del dict_stu["ram"]
print(dict_stu)
```

```python
# 6. Print all student names using keys().
print(dict_stu.keys())
```

```python
# 7. Print all student marks using values().
print(dict_stu.values())
```

```python
# 8. Print all student names with their marks using items().
for key,value in dict_stu.items():
    print(f"{key} marks are:{value}")
```

```python
# 9. Check if a specific student is in the dictionary.
print("Is yash is in dictionary :","yash" in dict_stu.keys())
```

```python
# 10. Check if a specific mark is present in the dictionary values.
print("Is 90 is in dictionary values : ",90 in dict_stu.values())
```

```python
# 11. Use get() to fetch a student's marks safely.
print("marks of vijay using get():",dict_stu.get("vijay"))
```

```python
# 12. Create a dictionary of employees with their departments.
dict_emp={"ajay":"HR","vijay":"Finance","sanjay":"IT","yash":"IT"}
```

```python
# 13. Access the department of a specific employee.
print("DEpt of yash is :",dict_emp["yash"])
```

```python
# 14. Add a new employee and department.
dict_emp["ram"]="Marketing"
print(dict_emp)
```

```python
# 15. Change the department of an existing employee.
dict_emp["ram"]="Sales"
print(dict_emp)
```

```python
# 16. Remove an employee from the dictionary.
del dict_emp["vijay"]
print(dict_emp)
```

```python
# 17. Sort the employee dictionary by employee names in ascending order.
print(dict(sorted(dict_emp.items(),key=lambda x:x[0])))
```

```python
# 18. Sort the employee dictionary by department names in ascending order.
print(dict(sorted(dict_emp.items(),key=lambda x:x[1])))
```

```python
# 19. Sort the employee dictionary by names in descending order.
print(dict(sorted(dict_emp.items(),key=lambda x:x[0],reverse=True)))
```

```python
## 20. Sort the employee dictionary by department in descending order.
print(dict(sorted(dict_emp.items(),key=lambda x:x[1],reverse=True)))
```

```python
# 21. Find all keys that have a specific value in a dictionary.
print("Employees in IT dept:",[key for key,value in dict_emp.items() if value=="IT" ])
```

```python
# 22. Count how many students scored above 80 marks.
cnt=0
for key,value in dict_stu.items():
    if value>80:
        cnt+=1
print("Number of students scored above 80 marks:",cnt)
```

```python
# 23. Create a nested dictionary for 3 students with age, grade, and department.
dict_nested={
    "ajay":{"age":20,"grade":"A","department":"IT"},
    "vijay":{"age":21,"grade":"B","department":"ENTC"},
    "sanjay":{"age":22,"grade":"C","department":"MECH"}
}
print(dict_nested)
```

```python
# 24. Access the grade of a student from the nested dictionary.
print("Grade  of ajays is :",dict_nested["ajay"]["grade"])
```

```python
# 25. Update the department of a student in the nested dictionary.
dict_nested["ajay"]["grade"]="O"
print("Grade  of ajays is :",dict_nested["ajay"]["grade"])
```

```python
# 26. Loop through the nested dictionary and print student names with their grades.
for key,value in dict_nested.items():
    print(f"{key} Are Scored grade is :{value["grade"]}")
```

```python
# 27. Merge two dictionaries of students into one.
dict_stu2={"rahul":75,"deepak":85}
dict_stu.update(dict_stu2)
print(dict_stu)
```

```python
# 28. Clear all entries in a dictionary.
del dict_stu2
print(dict_stu2)
```

```python
# 29. Copy a dictionary into a new dictionary and modify the copy.
dict_stu_copy=dict_stu.copy()
print(dict_stu_copy)
dict_stu_copy["ajay"]=95
print("Original Dictionary:",dict_stu)
print("Modified Copy:",dict_stu_copy)
```
