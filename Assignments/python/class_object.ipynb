```python
# Question 1
# Create a function that returns another function which multiplies a number by a fixed multiplier set during creation
def multiplier(num1):
    def inner(num2):
        return num1*num2
    return inner
result=multiplier(5)
print("the multplication  is :",result(3))
```

```python
# Question 2
# Build a class hierarchy for different types of bank accounts (Savings, Current) with a common withdraw method that behaves differently for each type
class account:
     def __init__(self,acc_id,balance):
          self.acc_id=acc_id
          self.balance=balance
     def withdraw(self,amount):
          if self.balance>=amount :
               self.balance=self.balance-amount
               print(f"The amount is withdraw{amount} and now the balance is :{self.balance}")

          else:
               print("We have insufficent balance")
          
     def deposit(self,amount):
          if amount>0:
               self.balance+=amount
          else:
               print("invalid amount")
class saving(account):
     def __init__(self,acc_id,balance,acc_type):
          super().__init__(acc_id,balance)
          self.acc_type=acc_type
          
          
     def withdraw(self, amount):
          if amount<=50000:
               return super().withdraw(amount)
          else:
               print("you limit to withdraw in one day is only 50000")
class current(account):
     def __init__(self,acc_id,balance,acc_type):
          super().__init__(acc_id,balance)
          self.acc_type=acc_type
     def withdraw(self, amount):
          if amount<=100000:
               return super().withdraw(amount)
          else:
               print("you limit to withdraw in one day is only 100000")

          
c1=saving(123,60000,"saving")
c2=current(124,1900000,"Current")
c1.withdraw(60000)
c2.withdraw(50000)
     
```

```python
# Question 3
# Write a function that generates an infinite sequence of Fibonacci numbers without storing all values in memory
def generator(ran):
    for i in range(1,ran):
        yield i

gen=generator(20)


```
