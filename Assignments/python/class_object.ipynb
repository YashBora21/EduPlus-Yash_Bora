```python
# Question 3
# Write a function that generates an infinite sequence of Fibonacci numbers without storing all values in memory
def fibonacci_generator():
    a, b = 0, 1
    while True:
        yield a
        a, b= b,a+b

gen = fibonacci_generator()
# Get first 20 Fibonacci numbers
for i in range(20):
    print(next(gen))


```

```python
# Question 4
# Create a timing decorator that measures and prints how long a function takes to execute
import time
def timer(func):
    def wrapper(*args,**kwargs):
        start=time.time()
        result=func(*args,**kwargs)
        end=time.time()
        print(f"the function {func.__name__} is run for this much time:{end-start}")
        return result
    return wrapper

@timer
def timer_ex(n):
    time.sleep(n)
    
timer_ex(4)

```

```python
# Question 6
# Implement a custom iterator class that returns squares of numbers from 1 to n

class Squares:
    def __init__(self, n):
        self.n = n
        self.i = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.i > self.n:
            raise StopIteration

        result = self.i * self.i
        self.i += 1
        return result
n = int(input("Enter n: "))

for sq in Squares(n):
    print(sq)
```

```python
# Question 8
# Create a logging decorator that prints the function name and arguments before execution
import time 
def login(func):
    def wrapper(*args, **kwargs):
        print(f"the function name is: {func.__name__} the args are {args} the kwargs are {kwargs}")
        result = func(*args, **kwargs)
        return result
    return wrapper

@login
def user(*args, **kwargs):
    time.sleep(2)
    print("doned")

user("yash", name="yash")
```

```python
# Question 10
# Design a Shape base class with area() method, and create Circle and Rectangle subclasses that implement it differently
class shape():
    def area(self): 
        pass
class circle(shape):
    def __init__(self,radius):
        self.radius=radius
    def area(self):
        print("the  area of the circle is :",3.142*self.radius*self.radius)
class rectangle(shape):
    def __init__(self,len,bre):
        self.len=len
        self.bre=bre
    def area(self):
        print("the area of the rectangle is :",self.len*self.bre)

c1=circle(3)
c2=rectangle(5,4)
c1.area()
c2.area()
```

```python
# Question 11
# Write a function that returns a function which maintains a running count of how many times it has been called
def outer():
    count=0
    def inner():
        nonlocal count
        count+=1
        return count
    return inner
result=outer()
print("the function call :",result())
print("the function call :",result())
print("the function call :",result())
```
