```python
# Question 4
# Create a timing decorator that measures and prints how long a function takes to execute
import time
def timer(func):
    def wrapper(*args,**kwargs):
        start=time.time()
        result=func(*args,**kwargs)
        end=time.time()
        print(f"the function {func.__name__} is run for this much time:{end-start}")
        return result
    return wrapper

@timer
def timer_ex(n):
    time.sleep(n)
    
timer_ex(4)

```

```python
# Question 13
# Implement a decorator that caches function results to avoid redundant calculations for the same inputs
import time
def caches(func):
    cache = {}
    def inner(*args):
        if args in cache:
            return cache[args]
        else:
            result = func(*args)
            cache[args] = result
            return result
    return inner
@caches
def calculation(*args):
    time.sleep(2)
    return sum(args)

print(calculation(2,3))
print(calculation(2,3))
print(calculation(4,5))
```

```python
# Question 14
# Build a generator that produces all prime numbers up to a given limit

def isprime(num):
    if num<2:
        return False
    for i in range(2,num):
        
        if num%i==0:
            return False
    else:
        return True

def generator(n):
    for i in range(n):
        if isprime(i):
            yield i

gen=generator(20)
for i in gen:
    print(i)
```

```python
# Question 19
# Create a decorator that converts the return value of a function to uppercase if it's a string

def decorator(func):
    def inner(str):
        if str.isalpha():
            result=func(str.upper())
            return result
            
        else:
            result=func(str)
            return result
    return inner

@decorator
def printer_is_str(str):
    return str

print(printer_is_str("yash"))
print(printer_is_str("y12"))
        
        
        
```

```python
def retry(func):
    def wrapper(*args, **kwargs):
        attempts = 3
        for attempt in range(attempts):
            try:
                return func(*args, **kwargs)
            except Exception as e:
                print(f"Attempt {attempt+1} failed: {e}")
                if attempt == attempts - 1:
                    raise
    return wrapper

@retry
def fis():
    print("Trying to execute fis()")
    raise ValueError("An error occurred!")

try:
    fis()
except Exception as e:
    print("Function failed after 3 attempts.")
```

```python
# Question 37
# Design a BankAccount class with private balance and public deposit/withdraw methods with transaction validation

class BankAccount:
    def __init__(self, initial_balance=0):
        self.__balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}. New balance: {self.__balance}")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if amount <= 0:
            print("Invalid withdrawal amount.")
        elif amount > self.__balance:
            print("Insufficient funds.")
        else:
            self.__balance -= amount
            print(f"Withdrew: {amount}. New balance: {self.__balance}")

    def get_balance(self):
        return self.__balance

# Example usage:
acct = BankAccount(100)
acct.deposit(50)
acct.withdraw(30)
acct.withdraw(200)
print("Final balance:", acct.get_balance())
```

```python
class FilterIterator:
    def __init__(self, iterable, condition):
        self.iterable = iterable
        self.condition = condition
        self.index = 0
    
    def __iter__(self):
        return self
    
    def __next__(self):
        while self.index < len(self.iterable):
            value = self.iterable[self.index]
            self.index += 1
            if self.condition(value):
                return value
        raise StopIteration

# Example usage
filtered = FilterIterator(list1, lambda x: x > 20)
for item in filtered:
    print(item)
```

```python
# Question 50
# Create an iterator that merges multiple sorted iterators into a single sorted sequence
class MergeIterator:
    def __init__(self, *iterables):
        self.data = sorted([item for it in iterables for item in it])
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= len(self.data):
            raise StopIteration
        value = self.data[self.index]
        self.index += 1
        return value


# Usage
a = [1, 3, 5]
b = [2, 4, 6]
c = [0, 7]

merged = MergeIterator(a, b, c)

for i in merged:
    print(i)
```
