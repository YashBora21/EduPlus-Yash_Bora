```python
n=int(input("Enter the Number of Students to Enroll"))

dict_stu={}
for i in range(n):
    courses=()
    
    name=""
    name=input("enter the name stdents")
    courses=input("enter the skills of courses").split(",")
    dict_stu[name]=courses
print(dict_stu)
```

1.	Identify all students enrolled in both ‘Data Science’ and ‘AI’. 

```python
for key,value in dict_stu.items():
    if "Data Science" in value and "AI" in value:
        print(key)
```

2.	Find the total number of unique courses offered.

```python
set1=set()
for key,value in dict_stu.items():
    set1.update(value)
print(set1)
```

3.	Find all courses that no student has repeated.

```python
dict_1={}
for key,value in dict_stu.items():
    for c in value:
        if c not in dict_1:
            dict_1[c] = 0
        dict_1[c]+=1
print(dict_1)
```
