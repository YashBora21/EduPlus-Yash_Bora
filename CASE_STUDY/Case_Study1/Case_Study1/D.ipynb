```python
n=int(input("Enter the number of the customer to enter"))
dict_cus={}
for i in range (n):
    name=input("Enter the customer name : ")
    product_list=input("Enter the list of products items where seperated by ,").split(',')
    product_list=tuple(product_list)
    dict_cus[name]=product_list
print(dict_cus)
```

1.	Display the total number of unique products bought by all customers combined.

```python
set1=set()
for key,value in dict_cus.items():
    set1.update(value)
print(set1)
```

2.	Find customers who purchased both ‘Laptop’ and ‘Headphones’.

```python
list1=[]
for key,value in dict_cus.items():
    if "laptop" in value and "headphones" in value:
        list1.append(key)
print(list1)
```

3.	Identify and display the most frequently purchased product(s) among all customers.

```python
frequent={}
for key,value in dict_cus.items():
    for i in value:
        if i not in frequent:
            frequent[i]=0
        frequent[i]+=1
print(frequent)
```

```python
max=0
product=""
for key,value in frequent.items():
    if int(max) < value:
        max=value
        product=key
print(f"the most sales item is :{product} ")
```
