```python
n=int(input("Enter The Number for days for which we have to put attendence: "))
dict_stu={}
for i in range(n):
    students=input(f"enter the day{i} present student:(seprated by ,) ").split(',')
    students=set(students)
    dict_stu[f"day{i}"]=students
print(dict_stu)
```

1.	Find students who were present every day.

```python
dict_count_day={}
for k,v in dict_stu.items():
    for c in v:
        if c not in dict_count_day:
           dict_count_day[c]=0
        dict_count_day[c]+=1
print(dict_count_day)
```

```python
for  k,v in dict_count_day.items():
    if v==n:
        print("the student who present all the days is:",k)
```

2.	Find students who were absent on at least one day.

```python
for  k,v in dict_count_day.items():
    if v!=n:
        print("the student who not  present all the days is:",k)
```

3.	Display the total unique students across all five days.

```python
set1=set()
for i in dict_stu.values():
    set1.update(i)
print("the unique students acorss all days :",set1)
```
