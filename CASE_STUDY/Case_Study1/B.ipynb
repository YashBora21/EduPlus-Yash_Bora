```python
n1=int(input("Enter the number of product of warehouses1"))
n2=int(input("Enter the number of product of warehouses2"))

warehouses1={}
warehouses2={}
for i in range(n1):
    porduct_id=int(input("enter the product id"))
    product_name=input("enter the product name")
    product_quantity=int(input(f"enter the {product_name} quantiy"))
    warehouses1[porduct_id]=product_name,product_quantity

for i in range(n2):
    porduct_id=int(input("enter the product id"))
    product_name=input("enter the product name")
    product_quantity=int(input(f"enter the {product_name} quantiy"))
    warehouses2[porduct_id]=product_name,product_quantity
print(warehouses1)
print(warehouses2)
```

```python
set1=set()

set2=set()
for key,value in warehouses1.items():
    set1.add(value[0])
for key,value in warehouses2.items():
    set2.add(value[0])
print(set1,set2)
```

1.	Display the common products available in both warehouses (based on product name).

```python
print(set1.intersection(set2))
```

2.	Display all unique products across both warehouses.

```python
print(set1.union(set2))
```
