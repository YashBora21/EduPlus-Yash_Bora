```python
n=int(input("Enter the number of the customer to enter"))
dict_cus={}
for i in range (n):
    name=input("Enter the customer name : ")
    movies_list=input("Enter the list of movies where seperated by ,").split(',')
    movies_list=tuple(movies_list)
    dict_cus[name]=movies_list
print(dict_cus)
```

2.	Display all unique movies in the database.

```python
set1=set()
for key,value in dict_cus.items():
    set1.update(value)

print(f"The total number of Unique movies watched by all customers are: {len(set1)}: {set1}")
```

3.	Identify movies that only one user has marked as favorite.

```python
dict_movie_fav={}
for j in set1:
    dict_movie_fav[j]=0

for j in range(n):
    for i in set1:
        a=input(f"Is this {i} is your favorite movie? (yes/no): ")
        if a=='yes':
            dict_movie_fav[i]+=1
    
        
```
