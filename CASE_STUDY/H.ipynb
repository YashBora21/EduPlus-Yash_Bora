```python
n=int(input("Enter the Number of Participants to enter"))

dict_Participants={}
for i in range(n):
    name=input("enter the name paticipant: ")
    sports_list=input("enter the list sports you have to enroll seperate by , :").split(",")
    dict_Participants[name]=sports_list
print(dict_Participants)
```

1.	Display participants who are taking part in both ‘Cricket’ and ‘Football’.

```python
list1=[]
for key,value in dict_Participants.items():
    if "cricket" in value and "football" in value:
        list1.append(key)
print(f"The list of participants who have enrolled both Cricket and Football are: {list1}")
```

2.	Find the total number of unique sports being played.

```python
set1=set()

for value in dict_Participants.values():
        set1.update(value)
print(set1)
```

3.	Create a tuple containing sports with the highest participation count.

```python
frequent={}
for key,value in dict_Participants.items():
    for i in value:
        if i not in frequent:
            frequent[i]=0
        frequent[i]+=1
print(frequent)
```

```python
max_freq=max(frequent.values())
freq_sport=()
for key,value in frequent.items():
    if value==max_freq:
        freq_sport+=(key,value)
print(f"The most popular sport is {freq_sport}")
```
