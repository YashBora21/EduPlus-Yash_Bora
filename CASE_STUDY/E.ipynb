```python
dict_stu_data={}
for i in range(0,5):
    day=input("Enter the day:")
    name=input(f"Enter the name students present on day:{day} ").split(',')
    name=set(name)
    dict_stu_data[day]=name
```

```python
print(dict_stu_data)
```

```python
dict_stu_present_days={}
for key,value in dict_stu_data.items():
    for i in value:
        if i not in dict_stu_present_days:
            dict_stu_present_days[i]=0
        dict_stu_present_days[i]+=1 
print(dict_stu_present_days)
```

1.	Find students who were present every day.

```python
list1=[]
for key,value in dict_stu_present_days.items():
    if value==5:
        list1.append(key)
print(f"The student list is present for all days: {list1}")
```

2.	Find students who were absent on at least one day.


```python
list2=[]
for key,value in dict_stu_present_days.items():
    if value!=5:
        list2.append(key)
print(f"The student list is absent for atleast one day: {list2}")
```

3.	Display the total unique students across all five days.

```python
set1=set()
for key,value in dict_stu_data.items():
    set1.update(value)
print(f"The total number of students in the class are: {len(set1)}: {set1}")
```
