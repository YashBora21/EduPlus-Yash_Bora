```python
import numpy as np
```

```python
mat1 = np.array([[1,2],[3,4]])
mat2 = np.array([[5,6],[7,8]])
print("\nthe matirx 1 is: \n",mat1)
print("\nthe matirx 2 is: \n",mat2)
print("\n matrix multiplication is :",np.dot(mat1,mat2))
print("\n matrix transpose is :",mat1.T)
print("\n matrix determinant  is :",np.linalg.det(mat1))
print("\n matrix inverse  is :",np.linalg.inv(mat1))

```

```python
orginal=np.array([1,2,3,4])
view_arr=orginal.view()
copy_arr=orginal.copy()
orginal[0]=99
print("\n the orinal arr is :",orginal)
print("\n the view arr is :",view_arr)
print("\n the copy arr is :",copy_arr)
```
