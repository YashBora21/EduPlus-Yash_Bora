```python
import numpy as np
```

```python
arr1=np.array([10,20,30])
arr2=np.array([
    [10,20,30],
    [40,50,60]
])
arr3=np.array([[
    [10,20,30],
    [40,50,60],
    [70,80,90]
]])
print("1d array is:\n",arr1)
print("2d array is:\n",arr2)
print("3d array is:\n",arr3)
```

```python
#array prop
print("dimension is:",arr3.ndim)
print("the size of arr3 is:",arr3.size)
print("the shape of arr3 is",arr3.shape)
print("the dtype of arr3 is",arr3.dtype)
```

```python
zeros=np.zeros((2,3))
ones=np.ones((3,3))
eye=np.eye(3)
rand=np.random.rand(2,3)
arrange_arr=np.arange(0,10,2)
linspace_arr=np.linspace(0,1,5)

print("\n the zeros array is :\n",zeros)
print("\n the ones array is :\n",ones)
print("\n the identical matrix array is :\n",eye)
print("\n the random   arry is :\n",rand)
print("\n the even  array is :\n",arrange_arr)
print("\n the linspace array is :\n",linspace_arr)
```

```python
print("\n the  1d array:\n",arr1)
print("\n 1st element in 1d array:\n",arr1[0])
print("\n 2st element in 1d array:\n",arr1[2])
print("\n 1t0 2 element in 1d array:\n",arr1[1:3])
print("\n 2 step element in 1d array:\n",arr1[::2])
```

```python
print("\n 2d array:\n",arr2)
print("\n 2,1 ele in 2d array:\n",arr2[1,2])
print("\n for 2 row 2 step 2d array:\n",arr2[:2,::2])
```

```python
#vstack and h stack
a=[[10,20],[30,40]]
b=[[50,60],[70,80]]
print("\n the h stacked array is:\n",np.hstack((a,b)))
print("\n the v stacked array is:\n",np.vstack((a,b)))
```
