```python
#class


class student:
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
    def details(self):
        print(f"name is : ",self.name)
        print(f"marks is : ",self.marks)
    def grade(self):
        if self.marks>=90:
            print("the grade is A+")
        elif self.marks>=70 and self.marks<90:
            print("the grade is B")
        elif self.marks>=50 and self.marks<70:
            print("the grade is C")
        elif self.marks>=40 and self.marks<50:
            print("the grade is B")
        else:
            print("you are fail")

s1=student("yash",90)
s2=student("prajwal",30)
s1.details()
s1.grade()
s2.details()
s2.grade()

```

```python
class bankaccount:
    def __init__(self,acc_no,balance):
        self.acc_no=acc_no
        self.balance=balance
    def deposit(self,amount):
        if amount>0:
            self.balance+=amount
            print("\nNow the total balance is : ",self.balance)
        else:
            print("\n Enter the valid amount")
    def withdraw(self,amount):
        if 0<amount<=self.balance:
            self.balance-=amount
            print("\n Now your balance is :",self.balance)
        else:
            print("\n insufficient Balance")
    def get_balance(self):
        return self.balance
    
account1=bankaccount("yash",5000)
account12=bankaccount("prajwla",10000)
account1.deposit(500)
account12.withdraw(500)
account12.withdraw(5000)
account1.withdraw(500)
account12.withdraw(5000)
print("\nnow the balance of the account is :",account1.get_balance())



```

```python
class vehicle:
    def __init__(self,brand,model):
        self.brand=brand
        self.model=model
    def start_engine(self):
        print(f"{self.brand}:{self.model}  engine is started")
    def stop_engine(self):
        print(f"{self.brand}:{self.model}  engine is stoppped")
    
class bike(vehicle):
    def __init__(self,brand,model,fuel_type):
        super().__init__(brand,model)
        self.fuel_type=fuel_type
    def details(self):
        print(f"Bike Brand:{self.brand}")
        print(f"model :{self.model}")
        print(f"fuel type:{self.fuel_type}")
        
class car(vehicle):
    def __init__(self,brand,model,cc):
        super().__init__(brand,model)
        self.cc=cc
    def details(self):
        print(f"Bike Brand:{self.brand}")
        print(f"model :{self.model}")
        print(f"fuel type:{self.cc}")
car1=car("Toyota","")


```
