```python
#Decorators
def logger(func):
    def wrapper():
        print(f"Running {func.__name__}...")
        func()
        
        print(f"{func.__name__} finshed")
    return wrapper
@logger
def greet():
    print("Hello,python learner")
greet()
@logger
def prajwal():
    print("hello  prajwal")
prajwal()
```

```python
#generator to simulate temperature
def temperature_stream():
    for temp in range(35,48):
        yield temp
def threshold_checker(threshold):
    def check(temp):
        return temp>threshold
    return check
#decorator
def logger(func):
    def wrapper(*args,**kwargs):
        print(f"checking {args[0]}degree exceeds threshold")
        result=func(*args,**kwargs)
        if result:
            print(f"Alert temprature")
        return result
    return wrapper
@logger
def analysis_data(temp):
    checker=threshold_checker(38)
    return checker(temp)
for t in temperature_stream():
    analysis_data(t)
```

```python
prices={"cisco":3400, "IBM":4500, "Microsoft":5000}
def genrator():
    for name, val in prices.items():
        yield  name,val

def threshold_checker(threshold):
    def check(value):
        return value > threshold
    return check

def logger(func):
    def wrapper(company,value):
        
        print(f"checking {company} price {value} exceeds threshold")
        result = func(company,value)
        if result:
            print("Alert price!!!!!!!")
        return result
    return wrapper

@logger
def analysis_data(company,value):
    
    checker = threshold_checker(4000)
    return checker(value)

for company,value in genrator():
    analysis_data(company,value)

```
