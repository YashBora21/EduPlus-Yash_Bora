```python
students={"id":101,"name":"Alice","age":23,"grade":"A"}
```

```python
new_students1={
    **{k:students[k] for k in list(students)[:2]},
    "city":"pune",
    **{k:students[k] for k in list(students)[2:]}
}
print(new_students1)
```
