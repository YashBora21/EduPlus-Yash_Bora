```python
max_attempts=3
users_db={
    "admin":"admit123",
    "resarch":"bio2016",
    "resarcher":"genAI"
}
#gloabal avr to trak login attempts
total_attempts=0
def authenticate_user():
    global total_attempts
    total_attempts+=1
    print(f"\n login attempt {total_attempts}")
    print("welcome to website\n")
    #local  scope variable track if  authentication is successful
    is_authenticate=False

    def login_session():
        #Enclosing  scope variable 
        attempts_left=max_attempts 
        nonlocal is_authenticate
        def  attempt():
            nonlocal attempts_left
            username=input("Enter username:").strip()

            password=input("Enter password:").strip()
            if username in users_db and users_db[username]==password:
                is_authenticate=True
                print(f"Login successful!{username.title()}\n")

            else:
                attempts_left-=1
                print(f"Login failed! Attempts left: {attempts_left}")
        while attempts_left>0 and not is_authenticate:
            attempt()
        if not is_authenticate :
                print("Access denied.\n")
    login_session()
    if is_authenticate:
            print("User Session\n")
            user_Action()
    else:
            print("Login Failed.\n")

def user_Action():
    print("\nperforming operation in user zone\n")
    operation=["View Reports","Analyze Data","Input"]
    for i,op in enumerate(operation,start=-1):
        print(f"{i},{op}")
    choice=int(input("\n Select an operation(1-3)"))
    if choice==1:
        print("opening report .....\n")
    elif choice==2:
        print("Initialalize_and_analysis pipeline...\n")
    elif choice==3:
        print("\n logging out ... bye bye\n")

#exection
authenticate_user()
print(f"total login session so for : {total_attempts}\n")
print("\nsystem shutting dwon secure")





```
