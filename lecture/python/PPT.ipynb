Dataset 1: Business Sales Records


```python
monthly_sales = [12000 , 15000 , 17000 , 14000 , 18000 , 16000 ,
20000 , 22000 , 21000 , 19000 , 23000 , 25000]
print("Monthly Sales Report")
print("---------------------")
print(monthly_sales)
```

10 Problem Statements

1 Highest sales recorded.

2 Lowest sales recorded.

3 Total yearly revenue.

4 Average monthly sales.

5 Total months tracked.

6 Sort sales (Ascending).

7 Sort sales (Descending).

8 Check for 20,000 mark.

9 Locate index of 18,000.

10 Filter months which sales Greater than  20,000.

```python
#1 Highest sales recorded
highest_sales = max(monthly_sales)
print("Highest Sales Recorded: $", highest_sales)
```

```python
#2 Lowest sales recorded.
lowest_sales = min(monthly_sales)
print("Lowest Sales Recorded: $", lowest_sales)
```

```python
#3 Total yearly revenue.
total_revenue = sum(monthly_sales)
print("Total Yearly Revenue: $", total_revenue)
```

```python
#4 Average monthly sales.
average_sales = total_revenue / len(monthly_sales)
print("Average Monthly Sales: $", average_sales)
```

```python
#5 Total months tracked.
total_months = len(monthly_sales)
print("Total Months Tracked:", total_months)
```

```python
#6 Sort sales (Ascending).
sorted_sales = sorted(monthly_sales)
print("Sales Sorted (Ascending):", sorted_sales)
```

```python
#7 Sort sales (Descending).
sorted_sales_desc = sorted(monthly_sales, reverse=True)
print("Sales Sorted (Descending):", sorted_sales_desc)
```

```python
#8 Check for 20,000 mark.
print("Is there any month with sales of $20,000 ? =", 20000 in monthly_sales)
    
```

```python
#9 Locate index of 18,000.
index_18000 = monthly_sales.index(18000)
print("Index of $18,000 sales:", index_18000)
```

```python
#10 Filter months with monthly sales greater than 20,000.
high_sales_months = [sale for sale in monthly_sales if sale > 20000]
print("Months with Sales Greater than $20,000:", high_sales_months)
```
