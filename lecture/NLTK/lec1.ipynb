```python
import nltk
nltk.download('movie_reviews')

```

```python
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize,sent_tokenize
text="Natural languages processing with NLTK. make it fun and learn it step by step."
sentences=sent_tokenize(text)
print("\n the snetences tokenization is:",sentences)
words=word_tokenize(text)
print(f"\nthe word tokeniation is :{words}")
```

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words('english'))
filter_word=[w for w in words if w.lower() not in stop_words]
print("the words main after toeknization is",filter_word)
```

```python
#stemming 
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
stemmer_words=[stemmer.stem(w) for w in filter_word]
print(stemmer_words)
```

```python
#lematization
nltk.download("wordnet")
from nltk.stem import WordNetLemmatizer
lemitizer=WordNetLemmatizer()
lemitizer_word=[lemitizer.lemmatize(w) for w in filter_word]
print(lemitizer_word)
```

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
pos_tags=nltk.pos_tag(filter_word)
print("part of speech :",pos_tags)
```

```python
import matplotlib.pyplot as plt
from nltk import FreqDist
freq_dist=FreqDist(filter_word)
print("word frequnecy:\n")
for word,freq in freq_dist.most_common(5):
    print(f"{word}:{freq}")

freq_dist.plot(10,title="top 10 used word")
plt.savefig("nltk.png")
plt.show()

```
