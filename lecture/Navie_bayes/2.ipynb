```python
import nltk
import random
from nltk.corpus import twitter_samples
nltk.download('twitter_samples')
```

```python
twiter_pos=twitter_samples.strings('positive_tweets.json')
twiter_neg=twitter_samples.strings('negative_tweets.json')
documnets=[(tweet,"pos") for tweet in twiter_pos]+[(tweet,"neg") for tweet in twiter_neg]
random.shuffle(documnets)
```

```python
documnets
```

```python
all_words=nltk.FreqDist(w.lower() for tweet in twiter_pos+twiter_neg  for w in tweet.split() )
words=list(all_words)[:2000]
```

```python
words
```

```python
def find_features(documnets):
    words=set(documnets)
    features={}
    for i in words:
        features[i]=(i in words)

    return features
feature_set=[(find_features(rev),category) for (rev,category)in documnets]
train_set,test_set=feature_set[100:],feature_set[:100]
classifier=nltk.NaiveBayesClassifier.train(train_set)
accuracy=nltk.classify.accuracy(classifier,test_set)*100
print("the classifier accuracy is: ",accuracy)
classifier.most_informative_features(20)
```
