```python
import tensorflow as tf  
print("version of the tensorflow :",tf.__version__)
print("GPU available :",tf.config.list_physical_devices('GPU'))
```

```python
scalar= tf.constant(42)
vector = tf.constant([1,2,3,4])
matrix=tf.constant([[1,-2],[2,-3]])
tensor_3d=tf.constant([[[1,2,3],[4,5,6]]])
print(f"\n scalar shape {scalar.shape},value :{scalar.numpy} :")
print(f"\n vector shape {vector.shape},value :{vector.numpy} :")
print(f"\n matrix shape {matrix.shape},value :{matrix.numpy} :")
print(f"\n tesnor_3d shape {tensor_3d.shape},value: {tensor_3d.numpy}")
```

```python
print(f"thensor dtype :{tensor_3d.dtype}")
print(f"tensor rank means dimensions of that variable :{tf.rank(tensor_3d).numpy()}")
print(f"tensor size :{tf.size(tensor_3d).numpy()}")
```

```python
a=tf.constant([[1,2],[3,4]])
b=tf.constant([[4,5],[5,6]])
print("the addition of matrix is:",tf.add(a,b))
print("the sub of matrix is:",tf.subtract(a,b))
print("the multply element of matrix is:",tf.multiply(a,b))
print("the   matrix multiplocation is:",tf.matmul(a,b))
```

```python
x=tf.constant([1.0,2.0,3.0])
print(f"the squre of matrix is :,{tf.square(x).numpy()}")
print(f"the squre root of matrix is :,{tf.sqrt(x).numpy()}")
print(f"the exponential  of matrix is :,{tf.exp(x).numpy()}")
```
