```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,200)
y=np.sin(x)
plt.plot(x,y)
plt.title("sin graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,c="blue",label="sin",linestyle='--',marker='o')
plt.plot(x,y2,c="red",label="cos",linestyle='-',marker='x')
plt.title("sin vs cos graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.legend()
plt.show()
```

```python
categories=['A','B','C','D']
values=[23,45,56,78]
plt.bar(categories,values,color="blue")
plt.title("bar chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend()
plt.show()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
colors=np.random.rand(50)
size=1000*np.random.rand(50)
plt.scatter(x,y,c=colors,sizes=size,cmap='viridis',alpha=0.5,label='bubbles')
plt.title("scatter plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.colorbar()
plt.legend()
plt.show()
```

```python
data=np.random.randn(100)
plt.hist(data,color="violet",edgecolor="black",bins=30,label="distribution",alpha=0.7)
plt.title("histogram  plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
lables=['apple','banana','grapes','mango']
sizes=[30,25,25,20]
plt.pie(sizes,labels=lables,autopct='%1.2f%%',shadow=True)
plt.title("histogram  plot")

plt.legend()
plt.tight_layout()
plt.show()
```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
plt.figure(figsize=(10,5))
#1st subplot
plt.subplot(1,2,1)#row column index
plt.plot(x,y1,c="blue",label="sin",linestyle='--',marker='o')
plt.title("sin(x)")
plt.legend()
plt.subplot(1,2,2)#row column index
plt.plot(x,y2,c="red",label="cos",linestyle='-',marker='x')
plt.title("cos(x)")
plt.legend()

plt.suptitle("subplot")
plt.tight_layout()

plt.savefig("sinvscos.png",dpi=300,bbox_inches='tight')
plt.show()
```
